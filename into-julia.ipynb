## Data structures
Julia has several built-in data structures, including arrays, tuples, dictionaries, and sets.

 **The indexing starts at 1, not 0.**

### Tuples
Tuples are ordered collections of elements that can be of different types. They are immutable, meaning that once created, their contents cannot be changed. You can create a tuple using parentheses `()`:

````julia
mytuple = (1, "Hello", π) # create a tuple with different types of elements
````

````julia
mytuple[3] # access the third element of the tuple
````

````julia
a, b, c = mytuple # unpack the tuple into separate variables
a
````

### Named tuples
Named tuples are similar to regular tuples, but they allow you to access their elements using names instead of indices. 

````julia
mynamedtuple = (number=1, greeting="Hello", pi=π) # create a named tuple with different types of elements
````

````julia
mynamedtuple.number # access the value associated with the name "number"
````

````julia
mynamedtuple[1]
````

### Dictionaries
Dictionaries are collections of key-value pairs. They are mutable, meaning that you can add, remove, or modify their contents. You can create a dictionary using the `Dict` constructor:

````julia
menu_caffe = Dict("Espresso" => 1.0, "Cappuccino" => 2.5) # create a dictionary with string keys and float values
````

````julia
menu_caffe["Espresso"] # access the value associated with the key "Espresso"
````

````julia
menu_caffe["Matcha Latte"] = 4.0 # add a new key-value pair to the dictionary
menu_caffe
````

````julia
myarray[1:3] # access the first three elements of the array
````

## Conditional statements


````julia
x = 3
y = 5
z = 10
if x < y && y < z
    println("x is less than y and y is less than z.")
elseif x < y || y < z
    println("x is less than y or y is less than z.")
else
    println("x is not less than y and y is not less than z.")
end
````

## Broadcasting
Broadcasting is a convenient syntax for applying any function over the elements of an iterable input. The result is a new iterable whose elements is the function application of the elements of the input.

````julia
h(x, y = 1) = x + y
````

````julia
x = [1, 2, 3]
h.(x)
````

````julia
h.(x,[1, 2, 3]) # component wise sum
````

# Multiple dispatch

- function: the name of the "function / process" we are referring to.
- method: what actually happens when we call the function with a specific combination of input arguments.

Dispatch means that when a function call occurs, the language decides somehow which of the function methods have to be used.

````julia
abstract type Animal end # this is an abstract type. a supertype of the below

struct Dog <: Animal   # this is a concrete type. a subtype of the above
    name::String
end

struct Cat <: Animal
    name::String
end
````

````julia
fido = Dog("Fido")
rex = Dog("Rex")
whiskers = Cat("Whiskers")
spotty = Cat("Spotty")
````

````julia
function encounter(a::Animal, b::Animal)
    verb = meets(a, b)
    println("$(a.name) meets $(b.name) and $verb")
end

meets(a::Animal, b::Animal) = "passes by"
````

````julia
encounter(fido, rex)
encounter(fido, spotty)
````

````julia
meets(a::Dog, b::Dog) = "sniffs"
meets(a::Dog, b::Cat) = "chases"
meets(a::Cat, b::Dog) = "hisses"
meets(a::Cat, b::Cat) = "slinks"
````

````julia
encounter(fido, rex)
encounter(fido, whiskers)
encounter(whiskers, spotty)
encounter(spotty, rex)
````
