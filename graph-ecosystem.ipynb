```julia
# create a simple graph with 5 vertices and 6 edges
g = DiGraph(5) # create an empty graph with 5 vertices
```

```julia
add_edge!(g, 1, 2) # add an edge from vertex 1 to vertex 2
add_edge!(g, 1, 3) # add an edge from vertex 1 to vertex 3
add_edge!(g, 2, 3) # add an edge
g
```

### Graph properties

```julia
println("Number of vertices: ", nv(g)) # number of vertices
println("Number of edges: ", ne(g)) # number of edges
println("Is the graph directed? ", is_directed(g)) # check if the graph is directed

println("Vertices: ", collect(vertices(g))) # list of vertices
println("Edges: ", collect(edges(g))) # list of edges

```

### Node and edge properties

```julia
println("In-degree of vertex 3: ", indegree(g, 3)) # in-degree of vertex 3
println("Out-degree of vertex 1: ", outdegree(g, 1)) # out-degree of vertex 1

println("Neighbors of vertex 1: ", collect(neighbors(g, 1))) # neighbors of vertex 1

println("Is there an edge from vertex 1 to vertex 2? ", has_edge(g, 1, 2)) # check if there is an edge from vertex 1 to vertex 2
myedges = collect(edges(g)) # get the list of edges
println("Source of edge (1, 2): ", src(myedges[1])) # source of the first edge
println("Destination of edge (1, 2): ", dst(myedges[1])) # destination of the first edge
```

# Jump.jl package
Jump.jl is a package for mathematical optimization in Julia. It provides a high-level interface for defining and solving optimization problems, including linear programming, integer programming, and nonlinear programming. It supports various solvers, including open-source solvers and commercial solvers like Gurobi and CPLEX.


```julia
Pkg.add("JuMP") # install the Jump.jl package
using JuMP # load the Jump.jl package
```

```julia
Pkg.add("HiGHS") # install the HiGHS solver, which is an open-source linear programming solver
using HiGHS # load the HiGHS solver
```

```julia
model = Model(HiGHS.Optimizer) # create a new optimization model using the HiGHS solver
@variable(model, x >= 0) # define a non-negative variable x
@variable(model, y >= 0) # define a non-negative variable y
@objective(model, Max, 3x + 2y) # set the objective function to maximize 3x + 2y
@constraint(model, x + y <= 4) # add a constraint that x + y must be less than or equal to 4
optimize!(model) # solve the optimization problem
assert_is_solved_and_feasible(model)
solution_summary(model)
println("Optimal value: ", objective_value(model)) # print the optimal value of the objective function
println("Optimal x: ", value(x)) # print the optimal value of x
println("Optimal y: ", value(y)) # print the optimal value of y
```

### Message passing

```julia
g = GNNGraphs.rand_graph(10,20)
g.ndata.x = ones(1,10)
GNNlib.propagate((xi,xj,e) -> xi.+1, g, +, xi = g.ndata.x) 
# xi.+1 is the message function, + is the aggregation function and xi = g.ndata.x is the initial node feature
```

```julia
g.graph # coo representation of the graph
```

The message passing procedure is the core of graph neural networks. It consists of three steps: message computation, message aggregation, and node update.
The `GraphNeuralNetworks.jl` propagate functtion implements the message computation and aggregation steps, which is then composed by `gather` and `scatter` functions. 

```julia
g.ndata.x # the node features 
```

```julia
messages = GNNlib.gather(g.ndata.x, g.graph[1]) # gather the node features of the neighbors of each node
```

```julia
message_function(x) = x + 1 # define a message function that adds 1 to the input
messages = message_function.(messages) # apply the message function to the gathered messages
```

```julia
new_messages = GNNlib.scatter(+, messages, g.graph[2]) # scatter the messages back to the nodes using the aggregation function +
```
